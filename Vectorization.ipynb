### One-Hot Encoding

```python
Corpos = ["dog eats meats man", "man eats meat"]
vocab = {}
count = 0
for doc in Corpos:
    for word in doc.split():
        if word not in vocab:
            count = count+1
            vocab[word] = count -1
print(vocab)
```

```python
def one_hot_encodding(doc):
    one_hot = []
    for word in doc.split():
        # print(word)
        temp = [0]*len(vocab)
        if word in vocab:
            print(vocab[word])
            temp[vocab[word]] =1
        one_hot.append(temp)
    return one_hot
```

```python
one_hot_encodding("dog eats meat")
```

```python
def BowRepresentation(all_uniquewords, documents):
    bow = dict.fromkeys(all_uniquewords,0)
    for word in documents:
        bow[word] = doc1.count(word)

    return bow
```

## TF-IDF

```python
import pandas as pd
import math
import sklearn
```

```python
first_sentence = "Data science is an Amazing career in the current world"
second_sentence = "Deep learning is a subset of machine learning"

first_sen_tokens = first_sentence.split(" ")
second_sen_tokens = second_sentence.split(" ")
print(first_sen_tokens), print(second_sen_tokens)
vacabulary = set(first_sen_tokens).union(set(second_sen_tokens))
vacabulary
```

```python
dict_1 = dict.fromkeys(vacabulary, 0)
dict_2 = dict.fromkeys(vacabulary, 0)

for word in first_sen_tokens:
    dict_1[word] +=1

for word in second_sen_tokens:
    dict_2[word] +=1

print(dict_1), print(dict_2)
```

```python
df = pd.DataFrame([dict_1, dict_2])
df
```

```python
## tf = freq of a word in a document / total number of words in a document
def calculateTF(wordDict, doc):
    tfdict = {}
    len_corpos = len(doc)

    for word, count in wordDict.items():
        tfdict[word] = count/len_corpos
    return tfdict

tf_sen_1 = calculateTF(dict_1, first_sen_tokens)
tf_sen_2 = calculateTF(dict_2, second_sen_tokens)

print(tf_sen_1), print(tf_sen_2)
```

```python
tf_dataframe = pd.DataFrame([tf_sen_1, tf_sen_2])
tf_dataframe
```

```python
import nltk
from nltk.corpus import stopwords
all_stop_words = stopwords.words("english")
```

```python
f1 = [word for word in first_sen_tokens if word not in all_stop_words]
f2 = [word for word in second_sen_tokens if word not in all_stop_words]
f1, f2
```

```python
def calculateIDF(doc):
    idfDict = {}
    len_doc = len(doc)

    idfDict = dict.fromkeys(doc[0],0)
    for word, val in idfDict.items():
        idfDict[word] = math.log10(len_doc/(float(val)+1))

    return idfDict
```

```python
idf = calculateIDF([dict_1 ,dict_2])
idf
```

```python
def computeTFIDF(tfBow, idfs):
    tfidf = {}
    for word, val in tfBow.items():
       tfidf[word] = val* idfs[word]

    return tfidf 
```

```python
idf1 = computeTFIDF(tf_sen_1, idf)
idf2 = computeTFIDF(tf_sen_2, idf)
```

```python
tfidf_df = pd.DataFrame([idf1, idf2])
tfidf_df
```
